#Remove those features that are'nt required.

```python
df.drop('Unnamed: 32',axis=1,inplace=True)
df.info()
```

# The diagnosis of breast tissues
*   **Benign** breast conditions refer to noncancerous changes or growths in the breast tissue that are not life-threatening but can sometimes increase the risk of developing breast cancer in the future.
*   **Malignant** tumor in the breast is a cancerous growth that can invade surrounding tissue and spread to other parts of the body

```python
#(M = malignant, B = benign)
df['diagnosis'].value_counts()
```

Encode the object data of the target data into numerical data for best model training.

```python
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df['diagnosis']=df.diagnosis.map({'M':1,'B':0})
df
```

```python
df['diagnosis'].value_counts()
```

# Split the data for Training & Testing

```python
x = df.drop('diagnosis',axis=1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x.shape)
print(y.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#Scale the features.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(x_train)
```

```python
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

#Model Selection
Select the model that is most preferable to your problem type.

```python
from sklearn.linear_model import LogisticRegression
lg_model=LogisticRegression()
```

```python
lg_model.fit(x_train,y_train)
```

```python
y_predicted=lg_model.predict(x_test)
```

```python
y_predicted
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predicted)*100
```
